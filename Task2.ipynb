```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
# Load the Dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
from sklearn.cluster import KMeans
k_ran = range(1,10)
sse = []

for k in k_ran :
  km = KMeans(n_clusters =k)
  km.fit(iris_df)
  sse.append(km.inertia_)
```

```python
sse
```

```python
# plotting to see by elbow method how many clusters we need
plt.title('The Elbow Method')
plt.plot(k_ran,sse)
plt.show()
```

```python
km1 = KMeans(n_clusters=3,max_iter=300 , random_state=10)
km1.fit(iris_df)
```

```python
y_means = km1.fit_predict(iris_df)
y_means
```

```python
# checking all of the centroids
km1.cluster_centers_
```

```python
iris_df = np.array(iris_df)
```

```python
plt.scatter(iris_df[y_means==1,2 ],iris_df[y_means==1,3 ], color='r' , label='Iris-setosa')
plt.scatter(iris_df[y_means==2,2 ],iris_df[y_means==2,3 ], color='b', label='Iris-virginica')
plt.scatter(iris_df[y_means==0,2 ],iris_df[y_means==0,3 ], color='g' , label='Iris-versicolor ')



plt.legend()
plt.show()
```
